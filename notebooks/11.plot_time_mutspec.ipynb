## MutSpec over time

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
path_all = "https://raw.githubusercontent.com/VoronkaAlexandr/Sars_Cov_2_MutSpec/main/Sars_Cov_2/new_data/data_obtained/MutChange_12spec_normed_one.csv"
path_syn = "https://raw.githubusercontent.com/VoronkaAlexandr/Sars_Cov_2_MutSpec/main/Sars_Cov_2/new_data/data_obtained/MutChange_12spec_SYN_normed_byone.csv"
path_ff  = "https://raw.githubusercontent.com/VoronkaAlexandr/Sars_Cov_2_MutSpec/main/Sars_Cov_2/new_data/data_obtained/MutChange_12spec_FF_normed_byone.csv"
```

```python
colors = [
    "#f8766d",
    "#de8c00",
    "#b79f00",
    "#7cae00",
    "#00ba38",
    "#00c08b",
    "#00bfc4",
    "#00b4f0",
    "#619cff",
    "#c77cff",
    "#f564e3",
    "#ff64b0",
]
colors = list(map(str.upper, colors))[::-1]
```

```python
def plot_time_mutspec_legacy(time_mutspec, outpath=None, label="", title=False):
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    time_mutspec.index = time_mutspec.index.str.replace(
        "20(\d\d)-(\d\d)_-_20(\d\d)-(\d\d)", 
        lambda x: "{}/{}-{}/{}".format(*x.group(2,1,4,3))
    )
    time_mutspec.plot(kind='bar', stacked=True, ax=ax, color=colors, edgecolor="white", linewidth=1.4, width=0.7)
    handles, labels = ax.get_legend_handles_labels()
    plt.rcParams['legend.title_fontsize'] = 14
    ax.legend(handles[::-1], labels[::-1], title='Mutations', loc='upper left', bbox_to_anchor=(1.02,1), fontsize=13)
    ax.set_xlabel("Time period", fontsize=14)
    ax.set_ylabel("Scaled substitution rate", fontsize=14)
    if title:
        ax.set_title("MutSpec by time, " + label);
    if outpath is not None:
        fig.savefig(outpath, bbox_inches="tight", pad_inches=.5)
    plt.show()


def plot_time_mutspec(time_mutspec, outpath=None):
    # Initialize a grid of plots with an Axes for each walk
    time_mutspec
    grid = sns.FacetGrid(time_mutspec, col="NucSubst", hue="NucSubst", palette="tab20",
                        col_wrap=3, height=2, aspect=2, sharey=False)

    # Draw a line plot to show the trajectory of each random walk
    grid.map(plt.plot, "variable", "value", marker="o")

    # # Adjust the arrangement of the plots
    # grid.tight_layout(w_pad=1)

    for ax in grid.axes.flat:
        for label in ax.get_xticklabels():
            label.set_rotation(90)

    grid.set_xlabels('')
    grid.set_titles('{col_name}')
    grid.set_ylabels('Percentage of\nsubstitution')
    grid.savefig('../figures/mutspec_by_time_all.pdf')
    if outpath:
        grid.savefig(outpath, bbox_inches="tight", pad_inches=.5)
    plt.show()
```

```python
time_mutspec_all = pd.read_csv(path_all, index_col=0)
time_mutspec_all['value'] *= 100

plot_time_mutspec(time_mutspec_all, "../figures/mutspec_by_time_all.pdf")
```

```python
time_mutspec_syn = pd.read_csv(path_syn, index_col=0)
time_mutspec_syn['value'] *= 100

plot_time_mutspec(time_mutspec_syn, "../figures/mutspec_by_time_syn.pdf")
```

```python
time_mutspec_syn4f = pd.read_csv(path_ff, index_col=0)
time_mutspec_syn4f['value'] *= 100

plot_time_mutspec(time_mutspec_syn4f, "../figures/mutspec_by_time_syn4f.pdf")
```

# Legacy

### Mut shares investigation

```python
x = np.arange(1, time_mutspec_all.shape[0] + 1)
y1 = time_mutspec_all["C>U"] / time_mutspec_all["U>C"]
y2 = time_mutspec_syn["C>U"] / time_mutspec_syn["U>C"]
y3 = time_mutspec_syn4f["C>U"] / time_mutspec_syn4f["U>C"]

params = {"s": 8}
plt.scatter(x, y1, label="all", **params)
plt.scatter(x, y2, label="syn", **params)
plt.scatter(x, y3, label="ff", **params)
plt.title("Changes of C>U and U>C ratio over time")
plt.xlabel("Time period")
plt.ylabel("Ratio [#C>U / #U>C]")
plt.legend()
plt.savefig("../figures/CU_UC_ratio.svg")
plt.show()
```

### Linear regression reports

1. all mutations
2. synonymous
3. four-fold

```python
X = sm.add_constant(x, prepend=False)

mod1 = sm.OLS(y1, X)
res1 = mod1.fit()

mod2 = sm.OLS(y2, X)
res2 = mod2.fit()

mod3 = sm.OLS(y3, X)
res3 = mod3.fit()
```
