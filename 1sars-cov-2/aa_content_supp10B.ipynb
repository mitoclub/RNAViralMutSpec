```python
import pandas as pd
import matplotlib.pyplot as plt
from Bio.Seq import Seq
from Bio.SeqUtils import seq3
import matplotlib.patches as patches
import seaborn as sns
from matplotlib.transforms import TransformedBbox
import numpy as np
from scipy.stats import mannwhitneyu
```

```python
pallete_zero=sns.color_palette("coolwarm", as_cmap=True)
```

```python
start_codons_data = pd.read_csv('../data/19.Codon_Start.csv')
end_codons_data = pd.read_csv('../data/19.Codon_End.csv')
```

```python
start_codons_data.head()
```

```python
table_for_heatmap = pd.DataFrame(columns=['codon', 'aa', 'start_value_mean', 'end_value_mean'])
```

```python
index_num = 0
for codon in start_codons_data.columns:
    codon_name = codon
    aa_name = Seq(codon).translate()[0]
    aa_name = seq3(aa_name)
    start_value = start_codons_data[codon].mean()
    end_value = end_codons_data[codon].mean()
    data = {'codon':codon_name, 'aa':aa_name, 'start_value_mean':start_value, 'end_value_mean':end_value}
    df = pd.DataFrame(data, index=[index_num])
    table_for_heatmap = table_for_heatmap.append(df)
    index_num+=1
```
