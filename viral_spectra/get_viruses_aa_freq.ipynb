```python
import pandas as pd
import os
import glob

from Bio import SeqIO
```

```python
from collections import Counter
```

# Get frequency of amino acids for all virus

```python
for root, dirs, files in os.walk('virus'):
    print('root =', root, 
          '\ndir =', dirs, 
          '\nfiles =', files)
```

```python
depth = 5
```

```python
path = '../viral_spectra/virus_raw_aminoacids'
all_path = []
for root,dirs,files in os.walk(path):
    if root[len(path):].count(os.sep) < depth:
        for f in files:
            all_path.append(os.path.join(root, f))
```

```python
# all_path
```

```python
all_path = [k for k in all_path if 'protein.faa' in k]
```

```python
all_path
```

```python
all_ref_seqs = [x.split('\\')[4] for x in all_path]
```

```python
all_taxid = [x.split('\\')[1] for x in all_path]
```

```python
list_of_seq = []

for path, ref_seq, taxid in zip(all_path, all_ref_seqs, all_taxid):
    
    fasta = list(SeqIO.parse(path, 'fasta'))
    
    for record in fasta:
        dictionary = {'taxid':taxid}
        dictionary.update({'refseq_id':ref_seq})
        dictionary.update(Counter(record.seq))
        list_of_seq.append(dictionary)
        # list_of_seq.append(dict({'taxid':11082}))
        # list_of_seq.append(Counter(record.seq))
```

```python
df = pd.DataFrame(list_of_seq)
```

```python
df_aminoacids = df.groupby(['taxid', 'refseq_id']).sum()
```

```python
df_aminoacids
```

```python
# df_aminoacids.sum(axis=1)
```

```python
df_aminoacids = df_aminoacids.div(df_aminoacids.sum(axis=1), axis=0)
```
