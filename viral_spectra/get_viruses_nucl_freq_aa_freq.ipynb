## For 4f codons

```python
# dict for four-fold codons
FOUR_FOLD_CODONS = {
    'GGT', 'GGC', 'GGA', 'GGG',  # Glycine (G)
    'CCT', 'CCC', 'CCA', 'CCG',  # Proline (P)
    'GCT', 'GCC', 'GCA', 'GCG',  # Alanine (A)
    'GTT', 'GTC', 'GTA', 'GTG',  # Valine (V)
    'ACT', 'ACC', 'ACA', 'ACG',   # Threonine (T)
    'CTT', 'CTC', 'CTA', 'CTG'    # Leucine (L) - only C-starting
}

def parse_location(location_str):
    #Parses str location and return a list of range
    if 'join(' in location_str:
        ranges = re.findall(r'(\d+)\.\.(\d+)', location_str)
        return [(int(start), int(end)) for start, end in ranges]
    else:
        start, end = re.findall(r'(\d+)\.\.(\d+)', location_str)[0]
        return [(int(start), int(end))]

def is_within_join(join_ranges, test_range):
    #Checks if test_range is inside any of join_ranges
    test_start, test_end = test_range[0]
    for (join_start, join_end) in join_ranges:
        if join_start <= test_start and test_end <= join_end:
            return True
    return False

def process_fasta_files(path_to_files):
    # Collect all records and location
    all_records = []
    join_records = []
    
    for file_path in path_to_files:
        refseq_id = os.path.basename(os.path.dirname(file_path))
        with open(file_path, 'r') as f:
            for record in SeqIO.parse(f, 'fasta'):
                # Extract location from description
                location_match = re.search(r'\[location=([^\]]+)\]', record.description)
                if location_match:
                    location_str = location_match.group(1)
                    location_ranges = parse_location(location_str)
                    all_records.append({
                        'refseq_id': refseq_id,
                        'record': record,
                        'location': location_ranges,
                        'is_join': 'join(' in location_str
                    })
                    if 'join(' in location_str:
                        join_records.append((refseq_id, location_ranges))
    
    # Filtered records: delete locations are included in the join records
    filtered_records = []
    for record_info in all_records:
        keep = True
        if not record_info['is_join']:
            for (refseq_id, join_ranges) in join_records:
                if record_info['refseq_id'] == refseq_id and is_within_join(join_ranges, record_info['location']):
                    keep = False
                    break
        if keep:
            filtered_records.append(record_info)
    
    return filtered_records

def count_four_fold_sites(sequence):
    #Count four-fold degenerate sites в последовательности"""
    counts = {'A': 0, 'T': 0, 'G': 0, 'C': 0}
    sequence = sequence.upper()
    
    for i in range(0, len(sequence) - 2, 3):
        codon = sequence[i:i+3]
        if len(codon) != 3:
            continue
        
        # Check 4f codon
        if codon in FOUR_FOLD_CODONS:
            third_nt = codon[2]
            if third_nt in counts:
                counts[third_nt] += 1
    
    return counts

def analyze_filtered_records(filtered_records):
    results = defaultdict(lambda: {'A': 0, 'T': 0, 'G': 0, 'C': 0})
    
    for record_info in filtered_records:
        refseq_id = record_info['refseq_id']
        sequence = str(record_info['record'].seq)
        counts = count_four_fold_sites(sequence)
        
        for nt, count in counts.items():
            results[refseq_id][nt] += count
    
    
    df = pd.DataFrame.from_dict(results, orient='index')
    df.index.name = 'refseq_id'
    return df.sort_index()
```

```python
filtered_records = process_fasta_files(all_path_nucl)
df_4f = analyze_filtered_records(filtered_records)
df_4f
```

```python
df_4f = df_4f.div(df_4f.sum(axis=1), axis=0)
df_4f
```

```python
df_4f = df_taxid_ref_seq.join(df_4f)
df_4f
```
